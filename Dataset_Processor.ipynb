```python
import os
import utils
import subprocess
import random
from subprocess import  check_output, CalledProcessError, STDOUT
import multiprocessing
from PIL import Image 
import numpy as np
```

```python
def cartoonletterbox(tup):
    f, fo, aspect_ratio, center_ratio = tup
    if not f.lower().endswith(('.jpg')):
        return
    try:
        im = Image.open(f)
        imagex = int(im.size[0])
        imagey = int(im.size[1])
        width = min(imagex, imagey*aspect_ratio) * center_ratio
        height = min(imagex/aspect_ratio, imagey) * center_ratio
        left =(imagex - width)/2
        top = (imagey - height)/2
        box = (left,top,left+width,top+height)
        im = im.crop(box)
        im = im.resize((256,256), Image.ANTIALIAS)
        im.save(fo, quality = 100)
    except:
        return None
    
def letterbox169(f, fo, aspect_ratio = 16./9., center_ratio = 1.0):
    if not f.lower().endswith(('.jpg')):
        return
    try:
        im = Image.open(f)
        imagex = int(im.size[0])
        imagey = int(im.size[1])
        width = min(imagex, imagey*aspect_ratio) * center_ratio
        height = min(imagex/aspect_ratio, imagey) * center_ratio
        left =(imagex - width)/2
        top = (imagey - height)/2
        box = (left,top,left+width,top+height)
        im = im.crop(box)
        im = im.resize((480,270), Image.ANTIALIAS)
        im.save(fo, quality = 100)
    except:
        return None
    
def letterbox(tup):
    f, fo, aspect_ratio, center_ratio = tup
    if not f.lower().endswith(('.jpg')):
        return
    try:
        im = Image.open(f)
        imagex = int(im.size[0])
        imagey = int(im.size[1])
        width = min(imagex, imagey*aspect_ratio) * center_ratio
        height = min(imagex/aspect_ratio, imagey) * center_ratio
        left =(imagex - width)/2
        top = (imagey - height)/2
        box = (left,top,left+width,top+height)
        im = im.crop(box)
        im.save(fo)
    except:
        return None
```

```python
datasetnames = ["YourName"] #folders containing datasets videos
extensions = ('.mp4', '.mkv') #acceptable video extensions
exclusions = ['YourName.mp4'] #original videos to not process, by filename
```

```python
#split all videos in each dataset folder into scenes, dump scenes into unified dest folder
for datasetname in datasetnames:
    videolist = os.listdir("data/videos/tgts/" + datasetname)
    for videoname in videolist:
        if not videoname.lower().endswith(extensions) or videoname in exclusions:
            continue
        command = ("scenedetect -i " 
                  + "data/videos/tgts/" + datasetname + "/" + videoname + " "
                  + "-o " + "data/videos/tgts_scenes/" + datasetname + " " 
                  + "detect-content split-video")
        print(command)
        os.system(command)
        utils.FrameCapture("data/videos/tgts/" + datasetname + "/" + videoname, 
                           "data/img_datasets/target_img_datasets/" + datasetname, 
                           prefix = videoname.split('.')[0],
                           div = 1, clearoutfolder = False)
```

```python
def processScene(tup):
    scene, datasetname = tup
    if not scene.lower().endswith(extensions): #skip junk files
        return
    inpath = "data/videos/tgts_scenes/" + datasetname + "/" + scene
    outfolder = "data/AnimeGAN_data/target_datasets/" + datasetname + "/" + scene
    if os.path.isdir(outfolder):
        return
    print("Splitting " + scene)
    utils.FrameCapture(inpath, outfolder, div = 24, startFrame = 3)

for datasetname in datasetnames:
    scenelist = os.listdir("data/videos/tgts_scenes/" + datasetname)
    
    try: #make dest folder if doesn't exist
        os.mkdir("data/AnimeGAN_data/target_datasets/" + datasetname)
    except OSError:
        print("Dataset folder probably already exists")   
    pool = multiprocessing.Pool()
    args = [(scene, datasetname) for scene in scenelist]
    pool.imap_unordered(processScene, args)
    pool.close()
    pool.join()
```

```python
#preprocess for cartoongan baseline
centerratio = 1.0
for datasetname in datasetnames:
    folder = "data/img_datasets/target_img_datasets/" + datasetname
    cganfolder = "data/img_datasets/target_img_datasets_CartoonGAN/" + datasetname
    try: #make dest folder if doesn't exist
        os.mkdir(cganfolder)
    except OSError:
        print("Dataset folder probably already exists")
        
    imagelist = os.listdir(folder)
    args = [(folder + "/" + image, cganfolder + "/" + image, 1., centerratio) for image in imagelist]
    pool = multiprocessing.Pool()
    pool.imap_unordered(cartoonletterbox, args)
    pool.close()
    pool.join()
```

```python
#generate target, resized and adjacent to edge promoting
def resize_promote(tup):
    folderin, folderout = tup
    print("Processing " + folderin)
    utils.cleardir(folderout)
    imagelist = os.listdir(folderin)
    framect = 0
    for image in imagelist:
        if not image.lower().endswith('.jpg'):
            continue
        letterbox169(folderin + "/" + image, folderout + '/' + image)
        framect += 1
    utils.edge_promoting_no_resize_inplace(folderout, framect)

centerratio = 1.0
for datasetname in datasetnames:
    folder = "data/AnimeGAN_data/target_datasets/" + datasetname
    pairfolder  = "data/AnimeGAN_data/paired_targets/" + datasetname
    try: #make dest folder if doesn't exist
        os.mkdir(pairfolder)
    except OSError:
        print("Dataset folder probably already exists")
        
    scenelist = os.listdir(folder)
    args = [(folder + "/" + scene, pairfolder + "/" + scene) for scene in scenelist]
#     resize_promote(args[1])
    pool = multiprocessing.Pool()
    pool.imap_unordered(resize_promote, args)
    pool.close()
    pool.join()
    
```

# Source dataset processor

```python
#https://stackoverflow.com/questions/3844430/how-to-get-the-duration-of-a-video-in-python
def getDuration(filename):
    command = [
        'ffprobe', 
        '-v', 
        'error', 
        '-show_entries', 
        'format=duration', 
        '-of', 
        'default=noprint_wrappers=1:nokey=1', 
        filename
      ]

    try:
        output = check_output( command, stderr=STDOUT ).decode()
    except CalledProcessError as e:
        output = e.output.decode()

    return output

def eraseWatermark(tup):
    videoname = tup
    if not videoname.lower().endswith(('.avi')):
        return
    vidpath = "data/videos/src/" + videoname
    outvidname = "data/videos/src/" + videoname.split('.')[0] + ".mp4"
    try:
        length = float(getDuration(vidpath))
    except:
        os.system("rm " + vidpath)
        return
    if length < 18.0:
        os.system("rm " + vidpath)
        return
    ccommand = "ffmpeg -i " + vidpath + " -ss 16 " + outvidname
    print(ccommand)
    os.system(ccommand)
    os.system("rm " + vidpath)
    
videolist = os.listdir("data/videos/src/")
pool = multiprocessing.Pool()
args = [videoname for videoname in videolist]
pool.imap_unordered(eraseWatermark, args)
pool.close()
pool.join()
```

```python
def processSrcScene(tup):
    videoname = tup
    if not videoname.lower().endswith(('.mp4')):
        return
    command = ("scenedetect -i " 
              + "data/videos/src/" + videoname + " "
              + "-o " + "data/videos/src_scenes" + " " 
              + "detect-content split-video")
    print(command)
    os.system(command)
videolist = os.listdir("data/videos/src/")
pool = multiprocessing.Pool()
args = [videoname for videoname in videolist]
pool.imap_unordered(processSrcScene, args)
pool.close()
pool.join()
```

```python
def splitSrcScene(tup):
    scene = tup
    if not scene.lower().endswith(extensions): #skip junk files
        return
    print("Splitting " + scene)
    inpath = "data/videos/src_scenes/" + scene
    outfolder = "data/AnimeGAN_data/src_data/" + scene
    frameoutfolder = "data/img_datasets/src_img/"
    utils.FrameCapture(inpath, outfolder, div = 24, startFrame = 3)
    utils.FrameCapture(inpath, frameoutfolder, prefix = scene.split('.')[0],
                       div = 1, clearoutfolder = False)

scenelist = os.listdir("data/videos/src_scenes/")
args = [scene for scene in scenelist]
pool = multiprocessing.Pool()
pool.imap_unordered(splitSrcScene, args)
pool.close()
pool.join()
```
